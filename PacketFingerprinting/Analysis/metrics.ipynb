```python
import numpy 
import matplotlib.pyplot as plt
```

```python
## Constants
BITS_INT = 16
BITS_FLOAT = 32
BITS_FLOAT64 = 64


BITS_per_feature = 32
BITS_per_iq = 32

Packet_bits = 512*BITS_per_iq
nr_features = range(1,64)
res = [Packet_bits/BITS_per_feature/i for i in nr_features]

plt.figure(figsize=(15, 4))
plt.plot(nr_features,res,label='train')
# plt.plot(nr_features,label='val')
# plt.title(k)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid(True)
plt.legend()
plt.show()
```
